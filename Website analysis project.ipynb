```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/pinma/Downloads/data-export.csv")
```

```python
df
```

```python
df.columns = df.iloc[0]
df = df.drop(index=0).reset_index(drop=True)
df.columns = ['Channel group','Datehour','Users','Sessions','Engaged sessions',	'Average engagement time per session',
              'Engaged sessions per user','Events per session','Engagement rate','Event count']
```

```python
df.head()
```

```python
df.info()
```

```python
df['Datehour'] =pd.to_datetime(df['Datehour'],format='%Y%m%d%H',errors='coerce')
df['Users'] = df['Users'].astype('int64')
df['Sessions'] = df['Sessions'].astype('int64')
df['Engaged sessions'] = df['Engaged sessions'].astype('int64')
df['Average engagement time per session'] = df['Average engagement time per session'].astype('float64')
df['Engaged sessions per user'] = df['Engaged sessions per user'].astype('float64')
df['Events per session'] = df['Events per session'].astype('float64')
df['Engagement rate'] = df['Engagement rate'].astype('float64')
df['Event count'] = df['Event count'].astype('int64')
df['Hours'] = df['Datehour'].dt.hour
```

```python
df['Hours'] = df['Datehour'].dt.hour
```

```python
df.info()
```

```python
df.head()
```

# sessions and user over time

```python
sns.set_theme(style='whitegrid')
fig, ax = plt.subplots(figsize=(10,5))
df.groupby('Datehour')[['Users','Sessions']].sum().plot(ax=ax)
plt.title('Sessions and users overtime',fontsize=14, fontweight='bold')
plt.xlabel('Datehour')
plt.ylabel('Count')
plt.show()
```

# Total users by channel

```python
sns.set_theme(style="whitegrid")

# Aggregate users by channel
channel_users = df.groupby('Channel group')['Users'].sum().sort_values(ascending=False)

# Create bar plot
fig, ax = plt.subplots(figsize=(10,5))
channel_users.plot(kind='bar', ax=ax, color='skyblue', edgecolor='black')

# Labels and title
ax.set_title("Website Traffic by Marketing Channel", fontsize=14, weight='bold')
ax.set_xlabel("Channel Group")
ax.set_ylabel("Number of Users")

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right')

# Add value labels on top of bars
for i, v in enumerate(channel_users):
    ax.text(i, v + 50, str(v), ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

# Average engagement time by channel

```python
plt.figure(figsize=(8,4))
sns.barplot(
    data=df,
    x='Channel group',
    y='Average engagement time per session',
    estimator=np.mean,
    hue='Channel group',        
    palette='magma',
    legend=False              
)
plt.title('Average Engagement Time per Session by Channel Group',fontsize=14, fontweight='bold')
plt.xlabel('Channel Group')
plt.ylabel('Average Engagement Time')
plt.xticks(rotation=45)
plt.show()
```

# Engagement Rate Distribution by channel

```python
plt.figure(figsize=(8,5))
sns.set_style("whitegrid")

sns.boxplot(
    data=df,
    x='Channel group',
    y='Engagement rate',
    hue='Channel group',       
    palette='coolwarm',
    legend=False
)

plt.title('Engagement distribution by channel group', fontsize=14, fontweight='bold')
plt.xlabel('Channel Group', fontsize=12)
plt.ylabel('Engagement Rate', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Engaged vs non engaged sessions

```python
session_df = df.groupby('Channel group')[['Sessions', 'Engaged sessions']].sum().reset_index()
session_df['Non-Engaged'] = session_df['Sessions'] - session_df['Engaged sessions']
session_df_melted = session_df.melt(id_vars='Channel group', value_vars=['Engaged sessions','Non-Engaged'])
plt.figure(figsize=(8, 5))
sns.barplot(data=session_df_melted, x='Channel group', y='value', hue='variable')
plt.title('Engaged vs Non-Engaged Sessions',fontsize=14, fontweight='bold') 
plt.xticks(rotation=45) 
plt.show()
```

# Traffic by hour and channel

```python
heatmap_data = df.groupby(["Hours", "Channel group"])["Sessions"].sum().unstack().fillna(0)
plt.figure(figsize=(12, 6)) 
sns.heatmap(heatmap_data, cmap="YlGnBu", linewidths=.5, annot=True, fmt='.0f') 
plt.title("Traffic by Hour and Channel",fontsize=14, fontweight='bold') 
plt.xlabel("Channel Group") 
plt.ylabel("Hour of Day") 
plt.show()
```

# Engagement rate vs sessions over time

```python
df_plot = df.groupby('Datehour')[['Engagement rate', 'Sessions']].mean().reset_index()
plt.figure(figsize=(10, 5))
plt.plot(df_plot['Datehour'],df_plot['Engagement rate'],label='Engagement rate',color='green')
plt.plot(df_plot['Datehour'],df_plot['Sessions'], label='Sessions', color='blue')
plt.title('Engagement Rate vs Sessions Over Time',fontsize=14, fontweight='bold')
plt.xlabel('Datehour') 
plt.legend() 
plt.grid(True) 
plt.show()
```
